```python
import helpers 
import config 
import numpy as np
```

```python
npz = np.load(config.RAW_DATA) 
X_train = npz["X_train"]
X_train
```

```python
import pandas as pd
Xtr_df = pd.DataFrame(X_train)
Xtr_df
```

```python
CAT = []
DISC = []
CONT = []

for col in Xtr_df.columns:
    vals = Xtr_df[col].dropna()
    n_unique = vals.nunique()
    dtype = vals.dtype

    # Heuristic rules
    if dtype == 'object':
        CAT.append(col)
    elif np.issubdtype(dtype, np.integer) and n_unique < 30:
        DISC.append(col)
    else:
        CONT.append(col)

print("Categorical:", CAT)
print("Discrete:", DISC)
print("Continuous:", CONT)
```
